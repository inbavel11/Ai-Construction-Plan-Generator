```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

np.random.seed(42)
n = 1000

df = pd.DataFrame({
    'area': np.random.randint(500, 4000, n),
    'num_rooms': np.random.randint(1, 6, n),
    'num_toilets': np.random.randint(1, 3, n),
    'num_floors': np.random.randint(1, 4, n),
    'building_type': np.random.randint(0, 2, n)  # 0: Residential, 1: Commercial
})

df['living_room'] = df['area'] * 0.25
df['kitchen'] = df['area'] * 0.10
df['bedroom_total'] = df['area'] * 0.40
df['bedroom_1'] = df['bedroom_total'] * 0.5
df['bedroom_2'] = df['bedroom_total'] * 0.3
df['bedroom_3'] = df['bedroom_total'] * 0.2
df['toilet_1'] = df['area'] * 0.05
df['toilet_2'] = df['area'] * 0.03

X = df[['area', 'num_rooms', 'num_toilets', 'num_floors', 'building_type']]
y = df[['living_room', 'kitchen', 'bedroom_1', 'bedroom_2', 'bedroom_3', 'toilet_1', 'toilet_2']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
test_input = [[1500, 3, 2, 1, 0]]  # 1500 sqft, 3 rooms, 2 toilets, 1 floor, residential
pred = model.predict(test_input)

for name, size in zip(y.columns, pred[0]):
    print(f"{name}: {int(size)} sqft")
```

```python
def normalize_room_sizes(room_sizes, total_area):
    current_total = sum(room_sizes.values())
    scaling_factor = total_area / current_total
    return {k: v * scaling_factor for k, v in room_sizes.items()}
```

```python
def convert_area_to_dimensions(room_sizes):
    layout = []
    for room, area in room_sizes.items():
        width = (area ** 0.5) * 1.2
        height = area / width
        layout.append({
            'room': room,
            'area': area,
            'width': round(width, 1),
            'height': round(height, 1)
        })
    return layout
```

```python
def fit_rooms(layout, max_width=50):
    x, y, row_height = 0, 0, 0
    for room in layout:
        if x + room['width'] > max_width:
            x = 0
            y += row_height + 1
            row_height = 0
        room['x'] = x
        room['y'] = y
        x += room['width'] + 1
        row_height = max(row_height, room['height'])
    return layout
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_plan(layout):
    fig, ax = plt.subplots(figsize=(10, 8))
    for room in layout:
        rect = patches.Rectangle(
            (room['x'], room['y']), room['width'], room['height'],
            edgecolor='black', facecolor='lightblue', lw=1.5
        )
        ax.add_patch(rect)
        ax.text(room['x'] + 0.5, room['y'] + room['height']/2,
                f"{room['room']}\n{int(room['area'])} sqft", fontsize=8)
    ax.set_xlim(0, max(r['x'] + r['width'] for r in layout) + 5)
    ax.set_ylim(0, max(r['y'] + r['height'] for r in layout) + 5)
    ax.set_aspect('equal')
    plt.title("Fitted Floor Plan Layout")
    plt.axis('off')
    plt.show()
```

```python
room_dict = {
    "living_room": 300,
    "kitchen": 100,
    "bedroom_1": 150,
    "bedroom_2": 130,
    "bedroom_3": 120,
    "toilet_1": 40,
    "toilet_2": 40
}

total_area = 1500

normalized_rooms = normalize_room_sizes(room_dict, total_area)

room_layout = convert_area_to_dimensions(normalized_rooms)

fitted_layout = fit_rooms(room_layout, max_width=50)

draw_plan(fitted_layout)
```
